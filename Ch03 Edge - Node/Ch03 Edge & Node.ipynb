### 有一4阶张量`img`其`shape=[10, 28, 28, 3])`，代表10张28*28像素的3通道RGB图像，问：

```python
import tensorflow as tf
import numpy as np

img = tf.random_uniform([10,28,28,3])

sess = tf.Session()
print(sess.run(img))
```

#### 1. 如何利用索引取出第2张图片？（注意：索引均从0开始，第二张则索引为1，下同）

```python
img2 = img[1]

img2.shape
```

#### 2. 如何利用切片取出第2张图片？

```python
img2_1 = tf.slice(img,[1,0,0,0],[1,28,28,3])

img2_1.shape
```

#### 3. 使用切片与使用索引取出的一张图片有何不同？

区别：使用索引得到的图片的rank为3,但是使用切片得到的shape与原始数据保持一致，均为4

#### 4. 如何取出其中的第1、3、5、7张图片？

```python
img_odd = img[0:8:2]

img_odd.shape
```

#### 5. 如何取出第6-8张（包括6不包括8）图片中中心区域（14*14）的部分？

```python
img_cen = img[6:8,7:21,7:21]

img_cen.shape
```

#### 6. 如何将图片根据通道拆分成三份单通道图片？

```python
path1 = tf.slice(img,[0,0,0,0],[10,28,28,1])

path2 = tf.slice(img,[0,0,0,1],[10,28,28,2])

path3 = tf.slice(img,[0,0,0,2],[10,28,28,3])

print(path1.shape)
print(path2.shape)
print(path3.shape)
```

#### 7.写出`tf.shape(img)`返回的张量的阶数以及`shape`属性的值。

```python
shape = tf.shape(img)

rank = tf.rank(shape) # rank is 1

sess = tf.Session()
print(sess.run(rank))

sess = tf.Session()  # result is [10 28 28 3]
print(sess.run(shape))
```
